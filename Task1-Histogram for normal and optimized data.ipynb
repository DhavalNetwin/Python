```python
import pandas as pd
import numpy as np
import datetime as datetime
import matplotlib.pyplot as plt
hsNData1 = pd.read_csv(r'E:\Dhaval\Elastic Search\Python\E-Book\Data File\highstoragesystem-data-for-energy-optimization\HRSS_normal_optimized.csv')
```

```python
#def clean_strings(strings):
result = []
cnt = 0
for index, value in enumerate(hsNData1['Timestamp']):
    if index == 0:
        result.append(cnt)
    elif index == len(hsNData1['Timestamp'])-1:
        result.append(cnt)
    elif hsNData1['Timestamp'][index] - hsNData1['Timestamp'][index - 1] < 0:
        cnt += 1
        result.append(cnt+1)
    else:
        result.append(cnt)
#return 
result
```

```python
hsNData1['sampleNum'] = result
len(result)
```

```python
def CreateHistogramChart(hsNData,W_O_WO_L,sampleNo,remove_index = 0):
    #hsNData = hsNData1
    #sampleNo = 1
    #W_O_WO_L = True
    fig = plt.figure()
    z = 1
    fig.set_figheight(20)
    fig.set_figwidth(20)
    #c.set_yticks()
    for i in hsNData:
        if i == 'Timestamp' or i == 'Labels' or i == 'sampleNum':
            z == z
        else:
            c = fig.add_subplot(6, 3, z)
            data = hsNData.loc[hsNData['sampleNum'] == sampleNo]
            data = data.loc[data[i] != remove_index]
            c.hist(data[i], bins=30, color=['g'], alpha=0.3)
            #print(set(data[i]))
            if W_O_WO_L == True:
                c.set_title(i)
                #c.set_ylabel('Count')
                c.set_yticks([])
            z += 1
    plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=2.0)
    fig.set_dpi(1000)
```

```python
CreateHistogramChart(hsNData1,True,1,0)
```
